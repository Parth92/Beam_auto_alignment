```python
import pykat
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=12)
```

# Setting up the model

The parameters are not correct. Needs to be updated.

```python
base = pykat.finesse.kat()
base.verbose = False
base.parse("""

# Input laser
# -----------------------------
l laser 1 0 n0   # Laser (Power = 1 W, wavelength offset = 0)
s s0 1 n0 nSM1a    # Space (Length = 1 m)

# Steering mirrors
# -----------------------------
bs SM1 1 0 0 45 nSM1a nSM1b nSM1c nSM1d  # Beam splitter (R=1, T=0, phi(tuning)=0, alpha=45)
s s1 1 nSM1b nSM2a
bs SM2 1 0 0 45 nSM2a nSM2b nSM2c nSM2d
s s2 1 nSM2b nM1a

# Cavity
# -----------------------------
m M1 0.99 0.01 0 nM1a nM1b  # cavity mirror1 (R=0.99, T=0.01, phi=0)
s lCav 1 nM1b nM2a
m M2 0.99 0.01 0 nM2a nM2b

# Setting RoCs
attr M1 Rc -0.6
attr M2 Rc 0.6

# Defning the cavity for spatial mode basis computation
cav cav1 M1 nM1b M2 nM2a

# Output
# -----------------------------
s sOut 0.1 nM2b nOut

# Photo diode
pd P nOut

# Beam camera
beam CCD nOut

# trace 2
""")
```

# Measuring transmitted power while scanning cavity

```python
# Copying base-object
kat = base.deepcopy()

# Adding simulation instructions for scanning the cavity
kat.parse("""
xaxis M2 phi lin -10 190 300   # set linear x axis as phi from -10 w_o to 190 w_o in 300 setps
yaxis abs  # set y axis as amplitude
maxtem 3  # max order of HG mode
""")

# Steering mirror misalignments [rad]
# ------------------------------------
b1x = 5e-5
b2x = 0
b1y = 0
b2y = 2e-5

kat.SM1.xbeta = b1x
kat.SM1.ybeta = b1y
kat.SM2.xbeta = b2x
kat.SM2.ybeta = b2y
# ------------------------------------

# Running simulation
out = kat.run()



# Plotting
fig = plt.figure(figsize = (7,5), dpi=100)
ax = fig.add_subplot(111)
ax.plot(out.x, out['P'], linewidth=2)
ax.grid()
ax.set_title(('$\\beta_{{1x}}$ = {} $\mu$rad,  $\\beta_{{2x}}$ = {} $\mu$rad, '+
              '$\\beta_{{1y}}$ = {} $\mu$rad,  $\\beta_{{2y}}$ = {} $\mu$rad').format(b1x*1e6, b2x*1e6, 
                                                                                      b1y*1e6, b2y*1e6))
ax.set_ylabel('Power [W]')
ax.set_xlabel('M2 tuning [deg]')
plt.show(fig)
```

# CCD camera

## Scanning only x-direction

```python
# Copying base-object
kat = base.deepcopy()

# Adding simulation instructions
kat.parse("""
xaxis CCD x lin -10 10 200
yaxis abs
maxtem 3
""")

# Steering mirror misalignments [rad]
# ------------------------------------
b1x = 5e-5
b2x = 0
b1y = 0
b2y = 2e-5

kat.SM1.xbeta = b1x
kat.SM1.ybeta = b1y
kat.SM2.xbeta = b2x
kat.SM2.ybeta = b2y
# ------------------------------------

# Cavity tuning
kat.M2.phi = 45

# Running simulation
out = kat.run()

# Plotting
fig = plt.figure(figsize = (7,5), dpi=100)
ax = fig.add_subplot(111)
ax.plot(out.x, out['CCD'], linewidth=2)
ax.grid()
ax.set_title(('$\\beta_{{1x}}$ = {} $\mu$rad,  $\\beta_{{2x}}$ = {} $\mu$rad, '+
              '$\\beta_{{1y}}$ = {} $\mu$rad,  $\\beta_{{2y}}$ = {} $\mu$rad').format(b1x*1e6, b2x*1e6, 
                                                                                      b1y*1e6, b2y*1e6))
ax.set_ylabel('Intensity [W/m$^2$]')
ax.set_xlabel('x [w$_0$]')
plt.show(fig)
```

## Scanning both x and y

```python
# Copying base-object
kat = base.deepcopy()

# Adding simulation instructions
kat.parse("""
xaxis CCD x lin -5 5 100
x2axis CCD y lin -5 5 100
yaxis abs
maxtem 5
""")

# Steering mirror misalignments [rad]
# ------------------------------------
b1x = 5e-5
b2x = 0
b1y = 0
b2y = 2e-5

kat.SM1.xbeta = b1x
kat.SM1.ybeta = b1y
kat.SM2.xbeta = b2x
kat.SM2.ybeta = b2y
# ------------------------------------

# Cavity tuning [deg]
kat.M2.phi = 45

# Running simulation
out = kat.run()

print(out.x)

# Plotting
FS = 15
fig = plt.figure(figsize = (7,5), dpi=100)
ax = fig.add_subplot(111)
x,y = np.meshgrid(out.x, out.y)
hej = ax.pcolormesh(x, y, out['CCD'], cmap=cm.binary_r)
ax.grid()
ax.set_title(('$\\beta_{{1x}}$ = {} $\mu$rad,  $\\beta_{{2x}}$ = {} $\mu$rad, '+
              '$\\beta_{{1y}}$ = {} $\mu$rad,  $\\beta_{{2y}}$ = {} $\mu$rad').format(b1x*1e6, b2x*1e6, 
                                                                                      b1y*1e6, b2y*1e6))
ax.set_ylabel('y [w$_0$]', size=FS)
ax.set_xlabel('x [w$_0$]', size=FS)
cbar = fig.colorbar(hej)
cbar.set_label(label='Intensity [W/m$^2$]', size=FS)
plt.show(fig)
```
